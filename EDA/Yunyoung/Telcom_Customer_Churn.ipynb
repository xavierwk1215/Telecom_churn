```python
plt.figure(figsize=(12,6))
colors = sns.color_palette('magma', n_colors=4)
ax = churn_data.AgeGroup.value_counts().reindex(age_labels).plot.barh(color=colors)

# - Teen : 19세
# - YoungAdult : 20세 ~ 29세
# - PrimeAdult :  30세 ~ 49세
# - MiddleAdult : 50세 이상 ~ 60세
# - Senior : 60세 이상 

age_group = ['Teen (19세 이하)', 'YoungAdult (20세 - 29세)','PrimeAdult (30세 - 49세)', 'MiddleAdult (50세 - 59세)', 'Senior (60세 이상)']
# y 축 텍스트 변경
ax.set_yticklabels(age_group, size = 12)

for i, v in enumerate(churn_data['AgeGroup'].value_counts().reindex(age_labels)):
    if i==0:
        plt.text(v, i, str(v), color='black', size = 12, weight = 'bold')
    else:
        plt.text(v-150, i, str(v), color='white', size = 12, weight = 'bold')

ax.grid(color='gray', linestyle='--', linewidth=0.5)
plt.xlabel('Count', fontsize=12)
plt.gca().invert_yaxis() 
plt.show()
```

```python
# 나이 범주에 따른 이탈 퍼센트 생성 
age_group_df = pd.crosstab(churn_data.AgeGroup, churn_data.Churn, values=churn_data.Churn,aggfunc='count')
counts = age_group_df.sum(axis=1)
age_group_df['PctNo'] = round(age_group_df['No'] / counts * 100)
age_group_df['PctYes'] = round(age_group_df['Yes'] / counts * 100)
age_group_df
```

```python
age_group_df[['PctNo', 'PctYes']].sort_index().plot(
    figsize=(12, 6),
    kind='barh',
    color=['#CED7EA', '#ff0f4f'],
    stacked=True
)
    
for i, v in enumerate(age_group_df['PctNo'].sort_index()):
    plt.text(v+(1-v)/2 , i, f"{int(v)}% ({age_group_df['No'][i]})", color='black', size = 12, weight = 'bold')



for i, v in enumerate(age_group_df['PctYes'].sort_index()):
    plt.text((100-v)+(v/3.5) , i, f"{int(v)}% ({age_group_df['Yes'][i]})", color='white', size = 12, weight = 'bold')

plt.legend(title='Churn', labels=['No', 'Yes'])
plt.title('Churn Percent of Age Group',fontsize=15, weight='bold' )
plt.show()
```

### 통신사 요금 관련 컬럼 
- Total Revenue 는 TotalCharges	, Total Extra Data Charges,	Total Long Distance Charges 합친 최종요금에 환불 금액을 제외한 최종 금액을 나타낸다
- Total Long Distance Charges와 Total Revenue 에 이상치가 확인된다
    - 주요 피쳐 선택 시 제외할지 고민해볼 필요가 있다  
- 'TotalCharges', 'Total Refunds', 'Total Long Distance Charges' 컬럼을 이용해 각 컬럼이 'Total Revenue' 차지하는 비중을 나타내는 컬럼 생성 : TotalChargePct, TotalLongDistancePct

```python
# Total 요금 관련 컬럼 추출 
charge_cols = [col for col in churn_data.columns if ('total' in str(col).lower())]
charge_cols
```

```python
charge_df = churn_data[charge_cols].copy()

charge_df['TotalChargePct'] = round( (charge_df['TotalCharges'] / churn_data['Total Revenue']) * 100 ,2)
charge_df['TotalLongDistancePct'] = round( (charge_df['Total Long Distance Charges'] / churn_data['Total Revenue']) * 100 ,2)

churn_data.loc[charge_df['TotalCharges'] == churn_data['TotalCharges'], 'TotalChargePct'] = charge_df['TotalChargePct']
churn_data.loc[charge_df['TotalCharges'] == churn_data['TotalCharges'], 'TotalLongDistancePct'] = charge_df['TotalLongDistancePct']


charge_df.loc[charge_df['TotalCharges'] == churn_data['TotalCharges'], 'Churn'] = churn_data['Churn']

charge_df.head(8)
```

 #### Total Revenue , Total Long Distance Charges 이상치 확인 
 IQR 을 이용해 계산 :  Q1 - 1.5 * IQR보다 작거나 Q3 + 1.5 * IQR보다 큰 경우

```python
def get_outliers(target_col): #이상치를 구하는 함수 
    Q1 = charge_df[target_col].quantile(0.25)
    Q3 = charge_df[target_col].quantile(0.75)
    IQR = Q3 - Q1

    outliers = charge_df[(charge_df[target_col] < Q1 - 1.5 * IQR) | (charge_df[target_col] > Q3 + 1.5 * IQR)]
    
    return outliers
```

앙상블을 이용한 RandomForest 


```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

params={
    'max_depth' : [6, 8, 10], # 결정 트리의 최대 깊이를 6, 8, 10 로 설정 
    'n_estimators' : [50, 100, 200], # 앙상블 내 사용될 결정 트리의 개수를 결정
    'min_samples_leaf' : [8, 12], # 리프 노드에 필요한 최소 샘플 수를 지정합니다. [8, 12] 값은 최소 샘플 수로 8 또는 12를 고려
    'min_samples_split' : [8,12], # 노드를 분할하기 위해 필요한 최소 샘플 수를 지정
}
```
